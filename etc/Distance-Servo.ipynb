```python
# import the GoPiGo3 drivers
import time
import easygopigo3 as easy

# program to move around, stop when in front of an object,
# try to determine the best position to turn and continue to move forward
"""
    180            0
         __dis__
        [       ]
       [[       ]]
        [  pwr  ]
         [  o  ]

 The distance sensor when the robot is in this orientation,
 will have these positions, where 0 is the distance sensor facing
 all the way to the right, and 180 is facing all the way left
"""

gpg = easy.EasyGoPiGo3()

# instantiate a servo, defaults to port I2C 1
servo = gpg.init_servo()
servo.reset_servo()

# instantiate distance sensor
dist_sens = gpg.init_distance_sensor()

# set initial speed for gpg
gpg.set_speed(300)
# start moving forward
gpg.forward()

while True:
    # keeps the robot moving forward each loop
    gpg.forward()
    
    # take readings while moving forward
    curr_dist = dist_sens.read_mm()
    # if an object is detected less than 150mm away, adjust gpg pos
    if curr_dist < 150:
        gpg.stop()
        
        # variable to keep track of servo position with lowest distance
        servo_lowest_pos = 0
        # adjustable variable to range servo position from 0 to 180 degrees
        servo_adjust = 0
        # use continuous distance readings for quicker readings
        dist_sens.start_continuous(100)

        # keep track of the highest reading
        highest_dist = curr_dist
        # split 180 degrees into 6 parts, range is 0..5 
        for i in range(5):
            # rotate servo to adjusted position
            servo.rotate_servo(servo_adjust)
            # get continuous distance sensor reading
            dist_read = dist_sens.read_range_continuous()
            
            # print("{}mm, {} degrees".format(dist_read, servo_adjust))
            
            # if the distance read is higher, then the gpg wants to move that way
            if dist_read > highest_dist:
                highest_dist = dist_read
                # print("new highest dist: " + str(highest_dist))
                # print(servo_lowest_pos)
                servo_lowest_pos = i
            # add 30 degrees to the adjustable server position
            servo_adjust += 30
        if servo_lowest_pos * 30 <= 90:
            # highest distance to the right
            # print("right, " + str(servo_lowest_pos*30))
            gpg.right()
            # sleep for 2 seconds to all allow the turn to take place
            time.sleep(2)
            # reset the servo to default position, 45 degrees
            servo.reset_servo()
        if servo_lowest_pos * 30 > 90:
            # highest distance found to the left
            # print("left, " + str(servo_lowest_pos*30))
            gpg.left()
            time.sleep(2)
            servo.reset_servo()
```
